```python
import pandas
import matplotlib.pyplot as plt
data = pandas.read_csv("reviews.csv", parse_dates = ['Timestamp']) #telling the dataframe to treat the date as date(int) format and not string 
from datetime import datetime
import matplotlib.pyplot
from pytz import utc

data.head()
```

### Rating average/count by day


```python
data['Day'] = data['Timestamp'].dt.date #It is for extracting the date only b
data.head()

day_average = data.groupby(["Day"]).count()
day_average
```

```python
plt.figure(figsize = (25,3))
plt.plot(day_average.index, day_average['Rating'])
```

```python
day_average['Rating'].max()
```

```python
day_average['Rating'].min()
```

### Rating average/count by week 


```python
data['Week'] = data['Timestamp'].dt.isocalendar().week # as mentioned in the warning when .week() was used 
data['Week'] = data['Timestamp'].dt.strftime('%Y - %U') # use %Y for the year , %m for the month 
data['Week']
```

```python
data
```

```python
week_average = data.groupby(['Week']).mean()
week_average.head()

plt.figure(figsize = (25,3)) # This determines the sizing of the graph 
plt.plot(week_average.index, week_average['Rating'])
```

### Average rating by month 


```python
data['Month'] = data['Timestamp'].dt.strftime('%Y - %m')
mon_average = data.groupby(['Month']).mean()
mon_average.head()
```

```python
data
plt.figure(figsize = (25,3))
plt.plot(mon_average.index,mon_average['Rating'])
```

```python
data_services = data.groupby(['Month']).mean()
data_services
```

### What day are the people happiest 


```python
data['Weekday'] = data['Timestamp'].dt.strftime('%A')
data
```

```python
weekday_aver = data.groupby(['Weekday']).count()
weekday_aver
```

```python
weekday_aver = data.groupby(['Weekday']).mean()
weekday_aver
plt.plot(weekday_aver.index, weekday_aver['Rating'])
```

```python
data['DayNum'] = data['Timestamp'].dt.strftime('%w') ##this %w is based on the google search 
weekday_aver = data.groupby(['Weekday','DayNum']).mean()
weekday_aver = weekday_aver.sort_values('DayNum')
weekday_aver.index
plt.plot(weekday_aver.index.get_level_values(0), weekday_aver['Rating']) # Here Index is having 2 set of values, we are asking the index to get the first level values ie: Names 
plt.plot(weekday_aver.index.get_level_values(1), weekday_aver['Rating'])
```
